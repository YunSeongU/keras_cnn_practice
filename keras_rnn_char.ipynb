```python
import tensorflow as tf
import numpy as np
```

```python
word  = "what character will come next" 
char_=list(set(word))
word2index = {w:i for i,w in enumerate(char_,1)}
word2index['<unk>'] = 0
word2index
```

```python
x_str = word[:-1]
y_str = word[1:]

x = [word2index[c] for c in x_str]  # char to index
y = [word2index[c] for c in y_str]  # char to index
embdin = len(x)
embdout = len(word2index)


index2word={i:w for w,i in word2index.items()}


x=np.array([x]) 
y=np.array([y])
```

```python

x= tf.convert_to_tensor(x, dtype=tf.float32)
print(x)
type(x)
```

```python
class MyModel(tf.keras.Model):
    def __init__(self):
        super(MyModel, self).__init__()
        self.embedding_layer = tf.keras.layers.Embedding(embdin,embdout)
        self.rnn = tf.keras.layers.SimpleRNN(embdout, return_sequences=True)
        self.fc = tf.keras.layers.Dense(embdout,activation=tf.keras.activations.softmax)
    
    def call(self,x):
        x = self.embedding_layer(x)
        x=self.rnn(x)
        x=self.fc(x)
        return x
    

model = MyModel()
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam', metrics=['accuracy'])
model.fit(x, y, epochs=200)
```

```python
predictions = model.predict(x, verbose=0)

```

```python

for i, prediction in enumerate(predictions):
    print(x[i].numpy())
    a__ = [index2word[j] for j in x[i].numpy()]
    print(a__)
    
    print()
    print()

    index = np.argmax(prediction, axis=1)
    print(index)
    result = [index2word[j] for j in index]
    print(''.join(result))
```
